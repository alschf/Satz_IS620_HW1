```python
adlist = mygraph2.adjacency_list()
```

```python
## find the node with largest degree
```

```python
largest = -1
nodeindex = -1
for x in range(0, len(adlist)):
    if len(adlist[x])>largest:
        largest = len(adlist[x])
        nodeindex = x
```
